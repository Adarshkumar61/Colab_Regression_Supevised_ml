```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
```

```python
calories_data = pd.read_csv('/content/calories.csv')
```

```python
exercise_data = pd.read_csv('/content/exercise.csv')
```

```python
exercise_data.head()
```

```python
combined_data = pd.concat([exercise_data, calories_data['Calories']], axis=1)
```

```python
combined_data.head()
```

```python
combined_data.isnull().sum()
#so there are no missing value is in our dataset
```

```python
combined_data.replace({'Gender': {'female': 0, 'male': 1}}, inplace= True)
```

```python
combined_data.head()
```

```python
combined_data.drop(['User_ID'], axis=1, inplace= True)
```

```python
combined_data.head()
```

```python
x = combined_data.drop('Calories', axis=1)
y = combined_data['Calories']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
scaler = StandardScaler()
scaler
```

```python
scaler.fit_transform(x_train)
scaler.transform(x_test)
```

```python
model = XGBRegressor()
model
```

```python
model.fit(x_train, y_train)
```

```python
prediction_on_x_train = model.predict(x_train)
```

```python
prediction_on_x_test = model.predict(x_test)
```

```python
score_r2 = r2_score(y_train, prediction_on_x_train)
score_r2
```

```python
score_r2 = r2_score(y_test, prediction_on_x_test)
score_r2
```

```python
score_mea = mean_squared_error(y_train, prediction_on_x_train)
score_mea
```

```python
score_mea = mean_squared_error(y_test, prediction_on_x_test)
score_mea
```
